```python
#Observamos qwue hay columnas innecesarias 3,5,7,9,11,13, se eliminan
columnas_inservibles = dataframe.columns[[3,5,7,9,11,13]]
dataframe=dataframe.drop(columns=columnas_inservibles)
dataframe
```

```python
#Ahora tambien se observa en el excel que hay demasiadas filas, en especial al final
#Estas contienen informacion util pero no esencial, asi que parfa evitar complicaciones serán eliminadas
#Definimos el rango a conservar desde la celda 1 hasta la 228 en vertical, lo demas será eliminado.

dataframe=dataframe.drop(range(231,275))
dataframe
```

```python
#un mejor metodo es
dataframe=dataframe.iloc[:230]
dataframe
```

```python
#eliminamos las primeras filas tambien
dataframe=dataframe.drop(range(0,7))
dataframe
```

```python
#Ahora podemos renombrar mas facilmente las columnas del dataframe
columnas_nuevas={'Back':'HDI_Rank','Table 1. Human Development Index and its components': 'Country','Unnamed: 2':'HDI_Index','Unnamed: 4':'Life_Expectancy',
                'Unnamed: 6':'Expected_schooling_years','Unnamed: 8':'Mean_schooling_years','Unnamed: 10':'GNI_per_capita','Unnamed: 12':'GNIRank_minus_HDIRank',
                'Unnamed: 14':'HDI_Rank_2020'}
dataframe.rename(columns=columnas_nuevas, inplace=True)
dataframe
```

```python
dataframe['Human_development_groups']=pd.NA
#Definimos la informacion a usar
filas_nuevas= [8,75,125,170,203]
#Ponemos las filas en las nueva columna de datos
for index in filas_nuevas :
    nuevo = dataframe.loc[index,'Country']
    dataframe.loc[index+1:,'Human_development_groups'] = nuevo
```

```python
#Hacemos una copia para poder contar la frecuencia de los diferentes estados: low,medium,high,very high y despues reemplazar esto en la columna
dataframe['copia']=dataframe['Human_development_groups']
mapper = {'Low Human development': 1,'Medium Human development': 2,'High Human development': 3,'Very High Human development': 4}
dataframe['copia']=dataframe['Human_development_groups'].replace(mapper)
dataframe
```

```python
#ahora si pueden ser eliminadas
dataframe=dataframe.drop(filas_nuevas)
```

```python
#Finalmente reemplazamos todas las comas con puntos en todas las filas excepto country, para asi homogenizar la convencion de usar el punto
dataframe['HDI_index']=dataframe.HDI_index.str.replace(',','.')
dataframe['Life_Expectancy']=dataframe.Life_Expectancy.str.replace(',','.')
dataframe['Expected_schooling_years']=dataframe.Expected_schooling_years.str.replace(',','.')
dataframe['Mean_schooling_years']=dataframe.Mean_schooling_years.str.replace(',','.')
dataframe['GNI_per_capita']=dataframe.GNI_per_capita.str.replace(',','.')
```
